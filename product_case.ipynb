В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?

Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?

### Решение

### Метрики для оценки успеха ежемесячных событий: 
    
Представим, что какое-то количество событий уже прошло, и у нас имеются накопленные исторические данные. Чтобы оценить разницу между событиями и их успешность, я бы работала со следующими метриками:

- **Installs** – количество установок
- **CAC** – сколько стоит привлечение нового пользователя в тематическое событие
- **ARPU** во время проведения событий и без них
- **ASD** – среднее время сессий
- **Retention** – насколько событие актуально в течение месяца для разных когорт пользователей

### Метрики для изменённых событий: 
Добавление неудачных попыток **изменит набор метрик** 

Эти метрики останутся такими же:
- **Installs** – количество установок
- **CAC** – сколько стоит привлечение нового пользователя в тематическое событие
- **ARPU** во время проведения событий и без них
- **ASD** – среднее время сессий
- **Retention** – насколько событие актуально в течение месяца для разных когорт пользователей

Добавятся:

- **Retention внутри уровней/Lose rate** – чтобы оценить, какой процент игроков возвращается и продолжает проходить событие после отката на несколько уровней назад
- **Lose conversion** - кол-во неудачных уровней / кол-во всех уровней. Чтобы оценить количество попыток прохождения

Другие метрики (например, время сессии и ARPU) позволят нам оценить, изменилось ли время, за которое игрок участвует в событии и влияет ли это на увеличение дохода. А метрики для уровней позволят понять поведение игрока во время прохождения события
