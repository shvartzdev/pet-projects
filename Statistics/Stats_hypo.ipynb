```python
import pandas as pd
import numpy as np
import seaborn as sns

import scipy.stats as stats
```

```python
#n подбрасываний, m экспериментов
n = 100
m = 1000

p = []

for i in range(m):
    sum = 0
    for j in range(1, n):
        rand = round(np.random.randint(0,2), 3)
        sum += rand
    p.append(sum/n)
print(p)
#print(sum)
#print(sum/n)

sns.histplot(x=p, bins=30)
```

```python
statistic, pvalue = stats.shapiro(p)
print(pvalue < 0.05)
```
