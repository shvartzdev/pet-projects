```python
import pandas as pd
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import scipy.stats as stats
```

```python

# Данные по выручке
data = np.random.normal(0, 1, 100)

# Расчет основных статистик
mean = np.mean(data)
std_dev = np.std(data, ddof=1)
n = len(data)


# Уровень доверия
confidence_level = 0.95
alpha = 1 - confidence_level

# Используем нормальное распределение для больших выборок
z_score = stats.norm.ppf(1 - alpha/2)
margin_of_error = z_score * (std_dev / np.sqrt(n))
ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error

print(f"Доверительный интервал (нормальное распределение): ({ci_lower}, {ci_upper})")

# Используем t-распределение для малых выборок
t_score = stats.t.ppf(1 - alpha/2, df=n-1)
margin_of_error_t = t_score * (std_dev / np.sqrt(n))
ci_lower_t = mean - margin_of_error_t
ci_upper_t = mean + margin_of_error_t

print(f"Доверительный интервал (t-распределение): ({ci_lower_t}, {ci_upper_t})")
```

```python
plt.hist(data)


#plt.axvline(x = mean, color='r', linestyle='-', label='mean', linewidth=2)
plt.axvline(x = std_dev, color = 'g', linestyle='--', label='std', linewidth=2)
plt.axvline(x = confidence_level, color='y', linestyle='-', label='confidence level', linewidth=2)
plt.axvline(x = alpha, color='y', linestyle='--', label='alpha', linewidth=2)

plt.axvline(x = ci_lower, color='b', linestyle='--', label='ci_lower', linewidth=2)
plt.axvline(x = ci_upper, color='b', linestyle='--', label='ci_upper', linewidth=2)

plt.axvline(x = ci_lower_t, color='r', linestyle='--', label='ci_lower_t', linewidth=2)
plt.axvline(x = ci_upper_t, color='r', linestyle='--', label='ci_upper_t', linewidth=2)


plt.legend()
```

```python
import numpy as np
import scipy.stats as stats

# Генерация больших выборок данных по выручке для двух групп
np.random.seed(0)
group_A = np.random.normal(loc=200, scale=50, size=1000)
group_B = np.random.normal(loc=210, scale=50, size=1000)

# Расчет основных статистик для каждой группы
mean_A = np.mean(group_A)
mean_B = np.mean(group_B)
std_A = np.std(group_A, ddof=1)
std_B = np.std(group_B, ddof=1)
n_A = len(group_A)
n_B = len(group_B)

# Уровень доверия
confidence_level = 0.95
alpha = 1 - confidence_level

# Расчет стандартной ошибки разницы между средними
std_error_diff = np.sqrt((std_A**2 / n_A) + (std_B**2 / n_B))

# Используем z-распределение для больших выборок
z_score = stats.norm.ppf(1 - alpha/2)

# Расчет доверительного интервала для разницы между средними
margin_of_error = z_score * std_error_diff
mean_diff = mean_B - mean_A
ci_lower = mean_diff - margin_of_error
ci_upper = mean_diff + margin_of_error

print(f"Среднее значение группы A: {mean_A}")
print(f"Среднее значение группы B: {mean_B}")
print(f"Разница между средними: {mean_diff}")
print(f"Доверительный интервал для разницы между средними: ({ci_lower}, {ci_upper})")
```

```python
#n подбрасываний, m экспериментов
n = 1000
m = 100

p = []

for i in range(m):
    sum = 0
    for j in range(n):
        rand = round(np.random.randint(0,2), 3)
        sum += rand
    p.append(sum/n)
#print(p)

sns.histplot(x=p, bins=30)
```

```python
statistic, pvalue = stats.shapiro(p)
print(statistic, pvalue, pvalue > 0.05)
```

```python
import matplotlib.pyplot as plt

mu = 0  # среднее
sigma = 1  # стандартное отклонение

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
pdf = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu)/sigma)**2)

plt.plot(x, pdf, label='PDF')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Normal Distribution PDF')
plt.legend()
plt.show()
```

```python
from scipy.stats import norm

mu = 0  # среднее
sigma = 1  # стандартное отклонение

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
cdf = norm.cdf(x, mu, sigma)

plt.plot(x, cdf, label='CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('Normal Distribution CDF')
plt.legend()
plt.show()
```
